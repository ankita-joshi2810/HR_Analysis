```python
# feature Engineering
# encoding - one hot encoding
one_hot_var = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
for i in one_hot_var:
  x_t = pd.get_dummies(df[i], prefix = i, prefix_sep='_', drop_first=True)
  df=df.join(x_t)
  df.drop(i,axis=1, inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
smote = SMOTE(sampling_strategy='minority')
x_sm,y_sm = smote.fit_resample(x,y)
```

```python
y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_sm,y_sm, test_size=0.3,random_state=24)
```

```python
from sklearn.tree import DecisionTreeClassifier
ml = DecisionTreeClassifier()
ml.fit(x_train,y_train)
y_pred = ml.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
